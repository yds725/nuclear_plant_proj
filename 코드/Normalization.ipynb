```
# 30.csv 전처리하기 (String variable 존재)
import pandas as pd
import numpy as np

df_30=pd.read_csv("/content/drive/My Drive/dacon(한수원)/output/30.csv")
print(len(df_30)) #580
print(df_30.shape[1]) #2141
```

```
# ON 일시 1로 대체, OFF 일시 0으로 대체 
for i in range(1,df_30.shape[1]):
   
    df_30.loc[df_30.iloc[:,i] == "OFF", df_30.columns[i]] = 0
    df_30.loc[df_30.iloc[:,i] == "ON", df_30.columns[i]] = 1
    
display(df_30)
```

```
df_30.to_csv("/content/drive/My Drive/dacon(한수원)/output/30.csv",index=False)
```
