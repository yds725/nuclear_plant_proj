```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_00=pd.read_csv("/content/drive/My Drive/output/0.csv")
len(df_00)
df_00.shape[1]
```

```


for i in range(1,df_00.shape[1]+1):
    q1,q2,q3 = np.percentile(df_00.iloc[:,i],[25,50,75])
    df_00.loc[df_00.iloc[:,i] <= q1, i-1] = 0
    df_00.loc[(df_00.iloc[:,i] > q1) & (df_00.iloc[:,i] <= q2),i-1] = 1
    df_00.loc[(df_00.iloc[:,i]> q2) & (df_00.iloc[:,i] <= q3),i-1] = 2
    df_00.loc[df_00.iloc[:,i] > q3,i-1] = 3
print(df_00)
```
